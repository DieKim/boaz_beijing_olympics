```python
# 나머지는 걍 따로 수집
def urlcrawler(date, url):
    
    #드라이버 정의
    driver = webdriver.Chrome('C:/Users/jauri/chromedriver.exe') 
    driver.get(url)
    driver.maximize_window()

    # 인기순 클릭
    driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[1]/div/span[2]/button')[0].click()
    time.sleep(1)

    # 더보기 클릭(한 번에 20개)
    driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[3]/button')[0].click()
    time.sleep(2)
    driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[3]/button')[0].click()
    time.sleep(2)
    driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[3]/button')[0].click()
    time.sleep(2)
    driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[3]/button')[0].click()
    time.sleep(2)

    # url 수집
    max_num = 100
    urls=[]
    for j in range(max_num):
        li = driver.find_elements_by_xpath('/html/body/div/div[2]/div[2]/div/div[2]/div[2]/div/div[2]/ul/li['+str(j+1)+']')[0]
        a = li.find_element_by_tag_name('a')
        urls.append(a.get_attribute('href'))

    # 데이터프레임 저장
    temp = pd.DataFrame()
    temp['date'] = ['2022-2-' + str(date)] * max_num
    temp['date'] = pd.to_datetime(temp['date'])
    temp['url'] = urls

    result = pd.DataFrame()
    result = pd.concat([result, temp], axis=0)
    result.to_csv('olympicsnews_urls_'+str(date)+'.csv', index=False)
    display(result)

    driver.quit()
    return result
```

### 2. news 크롤링

```python
def newscrawler(URL, driver):
    
    #(1) 준비
    import pandas as pd
    import numpy as np
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')        
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')  
    driver = webdriver.Chrome('C:/Users/jauri/chromedriver.exe', options=options) 
    result = pd.DataFrame()
    
    #(2) 수집
    URL2 = URL['url'].tolist()
    for i, url in enumerate(URL2):
        print('url open... {0}번째 뉴스 데이터를 수집합니다.'.format(i+1))
        print(url)
        driver.get(url)
        driver.maximize_window()
        time.sleep(1)
        
        #기사작성일자
        #news_date = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[1]/div/span[1]').text
        news_date = URL.iloc[i, 0]
        #기사주소
        news_url = url
        #기사제목                                                                 
        news_title = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[1]/h4').text  
        time.sleep(1)
        #기사헤드라인
        try:
            news_headline = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[3]/strong').text
        except:
            news_headline = np.nan
        time.sleep(1)
        #기사내용
        #news_content = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[3]').text
        #기자
        #news_reporter = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[3]/p[2]').text
        #언론사
        #news_press = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[3]/p[1]').text
        #좋아요
        #news_good = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[1]/ul/li[1]/a/span[2]').text 
        #슬퍼요
        #news_sad = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[1]/ul/li[2]/a/span[2]').text 
        #화나요
        #news_angry = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[1]/ul/li[3]/a/span[2]').text  
        #팬이에요
        #news_fan = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[1]/ul/li[4]/a/span[2]').text  
        #후속기사원해요
        #news_want = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[1]/ul/li[5]/a/span[2]').text  
        #추천수
        #news_recommend = driver.find_element_by_xpath('/html/body/div[2]/div[1]/div/div/div[1]/div/div[4]/div[2]/div[1]/div/a/em[2]').text
        
        #(3) 저장
        temp = pd.DataFrame()
        temp['news_date'] = [news_date]
        temp['news_url'] = [news_url]
        temp['news_title'] = [news_title]
        try:
            temp['news_headline'] = [news_headline]
        except:
            pass
#         temp['news_content'] = news_content
#         temp['news_reporter'] = [news_reporter]
#         temp['news_press'] = [news_press]
#         temp['news_good'] = [news_good]
#         temp['news_sad'] = [news_sad]
#         temp['news_angry'] = [news_angry]
#         temp['news_fan'] = [news_fan]
#         temp['news_want'] = [news_want]
#         temp['news_recommend'] = [news_recommend]
        result = pd.concat([result, temp], axis=0)
    
    driver.quit()
    return result
```

```python
driver = webdriver.Chrome('C:/Users/jauri/chromedriver.exe') 
URL = pd.read_csv('C:/Users/jauri/BOAZ/미니프로젝트_올림픽/olympics_urls_final.csv')
res = newscrawler(URL[1500:], driver)
res.to_csv('olympics_news_19_20.csv', index=False)
```
