```python
#최종 트윗 크롤링 함수
def tweetscrawler(date):
  query = '올림픽' #검색어
  #korea_geo = "%s,%s,%s" % ("35.95", "128.25", "1000km") # 대한민국 트윗 -> 어차피 검색어가 한국어니까 필요없을 듯
  until = '2022-02-'+str(date) #특정 일자 이전의 트윗(7-day limit) 
  num = 10000 #수집할 트윗 개수(1만개 기준 1시간 정도 걸림)

  tweet_list = []
  for status in tweepy.Cursor(api.search, q=query, until=until).items(num): 
      
      #트윗내용/트윗작성일자/리트윗수/좋아요수/작성자
      temp_list = [status.text, status.created_at, status.id_str, status.retweet_count, status.favorite_count]
      
      #트윗원본(리트윗한 글?)
      try: temp_list.append(status.retweeted_status.text)
      except: temp_list.append('Not Retweeted')

      #최종 트윗 리스트
      tweet_list.append(temp_list)

  df = pd.DataFrame(tweet_list, columns=['Text', 'Created_Date', 'User_Id', '#_of_Retweets', '#_of_Likes', 'Original_Text'])
  df.to_csv('/content/drive/MyDrive/boaz/올림픽트윗크롤링/olympics_tweets_02'+str(date-1)+'.csv', index=False)
  return df
```
